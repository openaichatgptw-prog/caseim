```python
# =========================================================
# + BLOQUE NUEVO: CRUCE DETALLE VENTAS → PRECIO CNH
# =========================================================

import duckdb
import pandas as pd

# Reabrir DuckDB (la tabla resultado_precios_lista ya existe en disco)
duck = duckdb.connect("precio_de_lista.duckdb")
print("🔄 DuckDB reabierto")

# Leer Excel Detalle Ventas
df_detalle = pd.read_excel(
    r"C:\Users\andres.hernandez\OneDrive - IMECOL S.A.S\Documentos\PRUEBA PI PRECIOS\Detalle Ventas.xlsx"
)

# Registrar DataFrame
duck.register("tmp_detalle", df_detalle)

# Cruce correcto por Referencia
duck.execute("""
CREATE OR REPLACE TABLE detalle_con_cnh AS
SELECT
    d.*,
    r."Precio Prorrateo" AS "Precio CNH"
FROM tmp_detalle d
LEFT JOIN resultado_precios_lista r
    ON UPPER(TRIM(d.Referencia)) = UPPER(TRIM(r.Referencia))
""")

# Exportar Excel final
df_out = duck.execute("SELECT * FROM detalle_con_cnh").df()


df_out.to_excel("Detalles Ventas Cruzado.xlsx", index=False)

duck.close()
print("✅ Detalle Ventas cruzado con Precio CNH generado correctamente")
```
